```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
from tqdm import tqdm


from full_dynamics import FullSimulation
```

```python
fig, axes = plt.subplots(1, 2, figsize=(3.36, 2.5), constrained_layout=True)

def f_w_prime(rho, c_0 = 1.0, c_1 = 1.0, c_2 = 1.0):
    return - c_1 + c_2 * rho

def W(rho):
    return (rho-0.0)**2 * (1.0 - rho)**2

rho = np.linspace(-0.5, 1.5, 1000)


# Example content for the first panel
axes[0].set_title("(a) First-Order Wetting", fontsize=8)
#axes[0].set_xlabel(r"$\rho_s$", fontsize=8)
axes[0].tick_params(axis='both', which='major', labelsize=8)

axes[0].plot(rho, -f_w_prime(rho, c_1=0.55, c_2=0.3), label=r"$T<T_W$", color='r', lw=1.5, alpha = 0.3, linestyle='--')
axes[0].plot(rho, -f_w_prime(rho, c_1=0.38, c_2=0.3), label=r"$T=T_W$", color='r', lw=1.5)
axes[0].plot(rho, -f_w_prime(rho, c_1=0.2, c_2=0.3), label=r"$T>T_W$", color='r', lw=1.5, alpha = 0.3, linestyle='dotted')
axes[0].plot(rho, np.sqrt(W(rho)), color='m', lw=1.5)
axes[0].text(0.1, 0.02, r"$\sqrt{W(\rho_s)}$", fontsize=9, color='m')
axes[0].text(-0.2, 0.45, r"$-f'_W(\rho_s)$", fontsize=9, color='r')
axes[0].set_xticks([0.0, 0.44, 1.0, 1.1])
axes[0].set_xticklabels([r"$\rho_v$", r"$\rho_{1}$",r"$\rho_l$ ", r"  $\rho_2$"], fontsize=9)
axes[0].scatter([0.43, 1.06], [0.246, 0.06], zorder=5, color='black')
rho_f = np.linspace(0.41, 1.05, 1000)
axes[0].fill_between(rho_f, np.sqrt(W(rho_f)), -f_w_prime(rho_f, c_1=0.38, c_2=0.3), color='lightgrey')
axes[0].set_xlim(-0.3, 1.5)
axes[0].set_yticks([])
axes[0].set_ylim(bottom=0, top=0.6)
axes[0].legend(fontsize=8)


# Example content for the second panel
axes[1].set_title("(b) Second-Order Wetting", fontsize=8)
#axes[1].set_xlabel(r"$\rho_s$", fontsize=8)
axes[1].tick_params(axis='both', which='major', labelsize=8)

axes[1].plot(rho, -f_w_prime(rho, c_1=1.0, c_2=1.5), label=r"$T<T_W$", color='r', lw=1.5, alpha = 0.3, linestyle='--')
axes[1].plot(rho, -f_w_prime(rho, c_1=1.5, c_2=1.5), label=r"$T=T_W$", color='r', lw=1.5)
axes[1].plot(rho, -f_w_prime(rho, c_1=2.0, c_2=1.5), label=r"$T>T_W$", color='r', lw=1.5, alpha = 0.3, linestyle='dotted')
axes[1].plot(rho, np.sqrt(W(rho)), color='m', lw=1.5)
axes[1].scatter([1.0], [0.0], zorder=5, color='black')
axes[1].set_xticks([0.0, 1.0])
axes[1].set_xticklabels([r"$\rho_v$", r"$\rho_l$"], fontsize=9)
axes[1].set_xlim(-0.3, 1.5)
axes[1].set_xlim(-0.3, 1.5)
axes[1].set_yticks([])
axes[1].set_ylim(bottom=0, top=1.0)
axes[1].legend(fontsize=8)

#plt.savefig('fig_Cahn.pdf', dpi=600, bbox_inches='tight')
plt.show()
```
