```python
data = np.load("data/J_Pe10.npz")
J_Pe10 = data["J_Pe10"]
Ls = data["Ls"]
J_x = np.load("data/example_PW_state.npz")["J_x"]
rho = np.flip(np.load("data/example_PW_state.npz")["rho"])
p = -np.flip(np.load("data/example_PW_state.npz")["p"])

plt.clf()
plt.rcParams.update({'font.size': 7})
fig, axs = plt.subplots(2, 2, figsize=(6.69, 2.5), height_ratios=[1, 4], width_ratios=[1, 2], dpi=600, sharex='col', gridspec_kw={'wspace': 0.6})


# Bottom-left panel (row 2, column 1)
ax1 = axs[1, 1]

# Plot the magnified J_x on the second y-axis (right)
#ax2 = ax1.twinx()
#line1, = ax2.plot(np.linspace(0, 10, 299), J_x, label=r'$J_x(x)$', color='blue')
#ax2.hlines(0, 0, 10, color='lightgrey', linestyles='dashed')
#ax2.tick_params(axis='y', labelcolor='blue')
#ax2.set_ylim(-0.03, 0.04)

# Create the left y-axis for p and rho
line4 = ax1.annotate(r'${J_{\mathrm{ss}}^{(\rho)}}$', xy=(4.0, 0.5), xytext=(3.9, 0.5),
                     fontsize=8, color='grey', ha='right', va='center')
line5 = ax1.annotate('', xy=(4.0, 0.5), xytext=(6.0, 0.5),
                     arrowprops=dict(facecolor='grey', edgecolor='none', shrink=0.05, width=2, headwidth=7, alpha=0.6), zorder=0, alpha=0.6)


ax1.hlines(0, 0, 10, color='lightgrey', linestyles='dashed', linewidth=1.0)
line2, = ax1.plot(np.linspace(0, 10, 300), p, label=r'$p(x)$', color='green')
line3, = ax1.plot(np.linspace(0, 10, 300), rho, label=r'$\rho(x)$', color='red')
#ax1.set_ylabel(r'$p(x), \rho(x)$')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylim(-1, 1)
ax1.set_xlabel(r'Spatial Position $x$')

ax1.set_xticks([0, 5, 10])
ax1.set_xticklabels(['-5', '0', '+5'])
x = np.linspace(4.5, 5.5, 300)
y = 0.2 * (1 + np.cos((x-5)*np.pi/0.5))
ax1.fill_between(x, 0*y-1, y-1, color='lightgrey')

# Combine all legends into one
lines = [line2, line3]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='lower left')
ax1.set_xlim(0, 10)

# Add panel label (b)
#ax1.text(0.07, 0.95, '(b)', transform=ax1.transAxes, fontsize=10, va='top', ha='right')

ax3 = axs[0, 1]
ax3.plot(np.linspace(0, 5, 150), rho[:150], color='red')
ax3.set_ylim(0.96, 1.0)
ax3.set_yticks([0.96, 0.98, 1.0])
#ax3.set_ylabel(r'$\rho(x)$')
ax3.set_title('(b) The steady-state current modifies the density profile')

# Add panel label (a)
#ax3.text(0.09, 0.9, '(a)', transform=ax3.transAxes, fontsize=10, va='top', ha='right')

# Combine the right-hand panels into a single plot
ax4 = axs[:, 0].flatten()  # Flatten the right-hand panels
for ax in ax4:
    ax.set_xticks([])
    ax.remove()  # Remove the individual axes



# Create a single combined plot for the right-hand panels
ax_combined = fig.add_subplot(1, 2, 1)
x = np.linspace(12, 22, 200)
y = 0.069*x**(-1) #+ 0.043*np.exp(-0.4*x)

ax_combined.plot(x, 0.062*x**(-1), zorder=1, color='green', linestyle='dashed')
#ax_combined.plot(x, 0.09*x**(-1), zorder=1, color='green', linestyle='dashed')
ax_combined.text(14.0, 0.0033, r"$\sim \ell_s^{-1}$", fontsize=7, color="green")

#ax_combined.plot(x, 0.069*x**(-1)+ 0.043*np.exp(-0.4*x), zorder=1, color='lightgrey', linestyle='dashed')

ax_combined.set_title('(a) The current decays with system size')
ax_combined.scatter(Ls[:-2], J_Pe10[:-2], marker='x', zorder=2, s=10, label=r'Simulation ($\text{Pe}=10$)')
ax_combined.scatter(np.load('data/J_Pe8.npz')['Ls'][:-2], np.load('data/J_Pe8.npz')['J_dx'][:-2], marker='x', zorder=2, s=10, label=r'Simulation ($\text{Pe}=8$)')
ax_combined.legend()
ax_combined.set_xlabel(r'System Size $\ell_s$')
ax_combined.set_ylim(2.5*10**(-3), 2*10**(-2))
ax_combined.set_xlim(7.5, 22.5)
ax_combined.set_ylabel(r'Steady-State Current  $J_{\mathrm{ss}}^{(\rho)}$')
ax_combined.set_yscale('log')
ax_combined.set_xscale('log')
ax_combined.set_xticks(Ls[:-2:4])  # Ensure tick positions are set
ax_combined.set_xticklabels(Ls[:-2:4])  # Match labels to tick positions

# Adjust layout for the 4-panel figure
plt.tight_layout()
plt.subplots_adjust(hspace=0.15, wspace=1.0)
#plt.savefig("Fig_current.pdf", bbox_inches='tight')
plt.show()
```
